# let's make data to test on

```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = [] #aka target data?
train_samples = [] # aka ?
```

```python
for i in range(50):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)

for i in range(1000):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
train_labels, train_samples = shuffle(train_labels, train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

# let's build a model

```python
#build model
import tensorflow as ts
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
#train model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

a sequential model is composed of 'dense layers'. 
- dense = densley connected = every node is connected
the 1st layer is determined by the input data
layers 2-3 are the layers that should basically look for features
layer 4 is the output data
- input_shape - determines the expected size of the input data
- units - i think this is like the number of nodes, arbitrary (?)
- activation - determines the function that follows the dense layer to scale the data i think
"softmax" so final layer will output the probability of either class/label (1 or 0)

```python
model = Sequential([
    Dense(units=16, input_shape=(1,), activation='relu'),
    Dense(units=32, activation='relu'),
    Dense(units=2, activation='softmax')
])
```

prints a visual summary of the model

```python
model.summary()
```

# let's train the model on our data

optimizer - combine loss function and model params to determine how the weights and biases are changed (think like the direction of steepest descent). learning_rate determines how much the function changes the weights each time (scales the gradient of descent)

loss - specifies the loss function, which calculates how accurate the model was for any given piece of data

metrics - the metrics we use to determine how successful a model was

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

fit function to process the data to be used in the model (recall the documentation from earlier)

x - scaled sample data (aka input data)

y - labels corresponding to scaled sample data (target data)

validation_split - what percent is held out of training set to be used for validation

batch_size - self explanatory, refers to samples per batch

epochs - how many times does the model run and adjust weights/biases

shuffle - true by default, good to remember. training/validation set is split BEFORE shuffle, it's good practice to shuffle beforehand

verbose - allows us to see output (0=none, 1=some, 2=all)

validation data - a set of labeled data that we check the accuracy of the model on (the model is not trained on this)

```python
model.fit(x=scaled_train_samples, y=train_labels, validation_split=0.1, batch_size=10, epochs=30, shuffle=True, verbose=0)
```

# let's predict some labels for data

```python
test_labels = []
test_samples = []
```

# plot your confusion matrix data

cm = 

```python
def plot_confusion_mat(cm, classes, normalize=False, 
                      title ='confusion matrix lol',
                      cmap = plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest',cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    if normalize:
        cm=cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print('normalized confusion matrix')
    else:
        print('confusion matrix, not normalized')
        
    print(cm)
    
    thresh = cm.max()/2.
    for i,j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                horizontalalignment="center",
                color="white" if cm[i,j] > thresh else "blue")
    plt.tight_layout()
    plt.ylabel("true label")
    plt.xlabel("predicted label")
```

```python
cm_plot_labels = ['no_symptoms', 'symptoms']
plot_confusion_mat(cm=cm, classes=cm_plot_labels, title='confusion matrix')
```
